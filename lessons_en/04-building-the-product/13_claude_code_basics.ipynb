## What is Claude Code?

Claude Code is a **command-line AI assistant** built by Anthropic. Key differences from ChatGPT:

| Feature | ChatGPT / Web Chat | Claude Code |
|---------|-------------------|-------------|
| Sees your code | No (you paste snippets) | Yes (reads entire project) |
| Edits files | No | Yes (directly modifies code) |
| Runs commands | No | Yes (tests, installs packages) |
| Understands context | Limited to conversation | Full project awareness |
| Verifies work | You check manually | Can run tests itself |

ChatGPT is like texting a consultant. Claude Code is like having the consultant **sit at your desk** with full access to your project.

## What's in CLAUDE.md?

Our CLAUDE.md contains:
1. **Commands** — how to run the project (`python output/chat.py`)
2. **Architecture** — how the system is organized (chat interface, pipeline, agents)
3. **Model choices** — which models and why (Claude Sonnet, Grok-4, Opus)

Claude Code reads this file first when you ask it to make a change. That's why it can make smart decisions about WHERE to put new code and HOW to follow existing patterns.

## The AI Development Workflow

Working with Claude Code follows 5 steps:

```
1. Understand -> 2. Plan -> 3. Implement -> 4. Verify -> 5. Iterate
```

Same pattern as the agent pipeline:

| AI Dev Workflow | Content Pipeline |
|----------------|------------------|
| Understand the problem | Research Agent |
| Plan the approach | Outline Agent |
| Implement the solution | Writer Agent |
| Verify the output | Human review |
| Iterate if needed | Edit and re-run |

### The steps:

1. **Understand** — Claude Code uses "plan mode" to explore the codebase before making changes. The Research Agent does the same — search first, then act.
2. **Plan** — Claude Code designs the approach and shows you for approval. The Outline Agent creates structure before writing.
3. **Implement** — Claude Code writes the code, creates files, modifies existing ones.
4. **Verify** — YOU review the changes using your knowledge from this course. Is the schema right? Is the model choice appropriate? Does it follow existing patterns?
5. **Iterate** — If something's off, tell Claude Code what to fix. Repeat until correct.

## Key Commands and Modes

| Command / Mode | What it does |
|----------------|-------------|
| `claude` | Start Claude Code in your project |
| Plan mode | Claude Code explores before coding — like the Research Agent |
| `/help` | See all available commands |
| `Shift+Tab` | Switch between plan mode and normal mode |

### Sub-agents

Claude Code delegates to specialized sub-agents for different tasks — the same idea as the **Agno Team** in `chat.py`. The leader (main Claude Code) decides what needs to happen, then delegates to specialists for file reading, code search, terminal commands, etc.

## MCPs — Connecting to Documentation

**MCP** (Model Context Protocol) servers give Claude Code access to external knowledge sources.

Example: Our project uses the **Agno framework**. Without an MCP, Claude Code only knows what's in our code files. With the Agno docs MCP connected:
- Claude Code can look up Agno API methods
- It knows the latest Agno features and patterns
- It can check if a function exists before using it

MCPs give Claude Code **reference books**. Without them, it works from memory. With them, it can look things up.

Practical example: "I want to add a new tool to the pipeline. With the Agno docs MCP connected, I can say 'add a Google Search Console tool using the Agno Toolkit pattern' and Claude Code knows the exact API."

## Prompting Claude Code Effectively

Everything from Lesson 06 (Prompts and Context) applies here.

### Bad prompt:
> "Fix the pipeline"

### Good prompt:
> "In output/pipeline.py, the writer agent sometimes generates articles under 1000 words. Add a word count check after the writer step — if the article is under 1000 words, re-run the writer agent with a prompt asking it to expand. Follow the existing error handling pattern with try/except."

The good prompt has:
- **Specific file** — `output/pipeline.py`
- **Clear problem** — articles under 1000 words
- **Specific solution** — add a word count check, re-run if needed
- **Constraints** — follow existing error handling pattern

### More examples:

| Task | Bad prompt | Good prompt |
|------|-----------|------------|
| Change writing style | "Make articles better" | "In output/agents/writer.py, update writer_agent instructions to write in a casual, conversational tone instead of formal" |
| Add a feature | "Add keyword tracking" | "Add a target_keyword_density field to ContentOutline in output/agents/schemas.py, and add instructions to the outline agent to calculate keyword density per section" |
| Debug | "It's broken" | "When running python output/chat.py and asking to create an article, the outline agent returns an error about missing fields. Check output/agents/schemas.py for schema mismatches" |

````python
# When you ask Claude Code to modify the pipeline, it reads this file first.
# You can do the same thing to understand what it's working with!

pipeline_path = os.path.abspath("../../output/pipeline.py")
with open(pipeline_path, "r", encoding="utf-8") as f:
    pipeline_code = f.read()

print(f"pipeline.py: {len(pipeline_code)} characters, {len(pipeline_code.splitlines())} lines\n")
print("First 40 lines:")
print("=" * 60)
for i, line in enumerate(pipeline_code.splitlines()[:40], 1):
    print(f"  {i:3d} | {line}")
````
