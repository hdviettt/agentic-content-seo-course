# Lesson 2: Lists and Dictionaries

## Lists — Ordered Collections

A **list** stores items in order, numbered starting from `0`.

```python
my_list = ["item 1", "item 2", "item 3"]
#            [0]        [1]        [2]
```

**Important:** Python counts from `0`, not `1`.

In our project, we use lists to store keywords, articles, sections, and more.

````python
keywords = ["SEO", "content marketing", "link building"]

# Print the whole list
print("All keywords:", keywords)

# Access items by position (index)
print("First keyword:", keywords[0])
print("Last keyword:", keywords[-1])

# Count items
print("Total keywords:", len(keywords))
````

## Adding and Removing List Items

- `list.append(x)` — Add `x` to the end
- `list.remove(x)` — Remove item `x`
- `list.insert(i, x)` — Insert `x` at position `i`

Lists can be changed after creation — this is called being **mutable**.

````python
keywords = ["SEO", "content marketing", "link building"]
print("Original:", keywords)

# Add new keywords
keywords.append("on-page SEO")
keywords.append("technical SEO")
print("After adding:", keywords)

# Remove a keyword
keywords.remove("link building")
print("After removing:", keywords)

# Insert at a specific position
keywords.insert(0, "Google ranking")
print("After inserting at front:", keywords)
````

## Dictionaries (dict) — Key-Value Pairs

A **dictionary** maps **keys** to **values**. Each piece of data has a name (key) and a value.

```python
contact = {
    "name": "Viet",        # key: "name", value: "Viet"
    "age": 20,             # key: "age", value: 20
    "company": "SEONGON"   # key: "company", value: "SEONGON"
}
```

In our project, each **article** is stored as a dict with keys like `title`, `status`, `word_count`, etc.

````python
article = {
    "title": "SEO Guide 2026",
    "status": "draft",
    "word_count": 1500,
    "author": "Viet"
}

# Access values by key
print("Title:", article["title"])
print("Status:", article["status"])
print("Words:", article["word_count"])

# Change a value
article["status"] = "published"
print("New status:", article["status"])

# Add a new key
article["views"] = 0
print("Full article:", article)
````

## Nested Structures — Data Inside Data

In practice, data is often more complex. A dict can contain lists, a list can contain dicts, and so on.

Example: an article has multiple sections, and each section is a dict inside a list.

This is how our AI agent project organizes article data.

We'll build this up in two steps:
1. First: a **list of dicts** (like a spreadsheet)
2. Then: a **dict containing lists and dicts** (the full pattern used in our project)

````python
# A list of dictionaries — like a spreadsheet where each row is a dict
articles = [
    {"title": "SEO Basics", "status": "published", "views": 1200},
    {"title": "Link Building", "status": "draft", "views": 0},
    {"title": "Content Strategy", "status": "published", "views": 800},
]

# Access a specific article (list index, then dict key)
print("First article:", articles[0]["title"])
print("Its status:", articles[0]["status"])
print(f"Total articles: {len(articles)}")
````

````python
# Now the full picture: a dict containing a list of dicts
# This is exactly how our project stores article data
article = {
    "title": "On-Page SEO Guide 2026",
    "author": "Viet",
    "keywords": ["on-page SEO", "optimization", "Google ranking"],
    "sections": [
        {"heading": "Title Tags", "word_count": 500},
        {"heading": "Meta Descriptions", "word_count": 400},
        {"heading": "Header Tags", "word_count": 600},
    ]
}

# Access nested data step by step
print(f"Article: {article['title']}")
print(f"First keyword: {article['keywords'][0]}")
print(f"First section: {article['sections'][0]['heading']}")
print(f"Its word count: {article['sections'][0]['word_count']}")
````

## `for` Loops — Iterating Over Items

A **`for` loop** runs code once for **each item** in a list.

```python
for item in my_list:
    # do something with item
```

**Note:** Code inside the loop must be **indented** by 4 spaces. This is a strict Python rule.

### `enumerate()` — Getting Position Numbers

When you need both the **position number** and the **item** in a loop, use `enumerate()`:

```python
for i, keyword in enumerate(keywords, 1):
```

This gives you two values each loop:
- `i` — the position number (starting from 1, because we passed `1`)
- `keyword` — the item itself

Without `enumerate`, you'd only get the item and would have to track the number yourself.

````python
keywords = ["on-page SEO", "content marketing", "link building", "technical SEO", "local SEO"]

print("=== Keyword List ===")
print()

for i, keyword in enumerate(keywords, 1):
    print(f"  {i}. {keyword}")

print()
print(f"Total: {len(keywords)} keywords")

# Filter keywords containing "SEO"
print()
print("SEO-related keywords:")
for keyword in keywords:
    if "SEO" in keyword:
        print(f"  -> {keyword}")
````
