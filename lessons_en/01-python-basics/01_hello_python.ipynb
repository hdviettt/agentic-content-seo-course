## Where no-code stops working

You could build a basic content pipeline in n8n. AI node → Airtable node → trigger → done. Simple automations? n8n is faster than code. No argument.

But here's where it breaks:

**Batch speed.** n8n runs items one at a time. No parallel execution — that's a platform limit, not a bug. 20 articles × 5 min each = 100 minutes of waiting. With code, you run 3-5 at once. Same batch, ~30 minutes.

**Structured output.** n8n's output parser is unreliable. Malformed JSON, validation errors after a few turns, breaks on Markdown inside strings. Open GitHub issues, some 2+ years old. With code, you force the AI to output valid data at generation time. It doesn't "usually work." It always works.

**Conversation.** In n8n, every workflow run starts from zero. You can't say "that article failed, retry it" — there's no "it." You find execution #47 on the dashboard and click retry. With code, you build a system that remembers context. You talk to it. It acts.

These aren't things n8n will fix in an update. They're limits of how visual workflow builders work.

n8n got you here. Code gets you further.

## Variables — Labeled Boxes

A **variable** stores data under a name. You pick the name, assign the data with `=`.

The `=` sign means "assign a value" — not the same as "equals" in math.

````python
name = "Viet"
print(name)

age = 20
print(age)
````

## Strings and f-strings — Inserting Variables into Text

A **string** is text data, always wrapped in quotes: `"..."` or `'...'`.

An **f-string** lets you insert variable values into a string. Just add `f` before the opening quote and put variables inside `{}`.

```python
f"Hello {variable_name}"
```

This pattern is used everywhere in our project's codebase.

````python
company = "SEONGON"
role = "AI Leader"

print(f"{name} works at {company}")
print(f"Role: {role}")
print(f"{name} - {role} @ {company}")
````

## Numbers — Basic Math

Python can do math like a calculator:

| Operation      | Symbol | Example       |
|----------------|--------|---------------|
| Addition       | `+`    | `5 + 3` = 8   |
| Subtraction    | `-`    | `10 - 4` = 6  |
| Multiplication | `*`    | `3 * 7` = 21  |
| Division       | `/`    | `20 / 4` = 5.0|

You can combine numbers with f-strings for readable output.

````python
articles = 5
words_per_article = 2000
total_words = articles * words_per_article

print(f"Articles: {articles}")
print(f"Words per article: {words_per_article}")
print(f"Total words: {total_words}")
print(f"If you write {articles} articles/week, that's {articles * 4} per month!")
````
