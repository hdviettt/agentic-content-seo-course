## Prompt là gì?

Mọi thứ bạn gửi cho LLM đều là **prompt**. Chỉ vậy thôi. Không có phép thuật gì cả -- đó chỉ là đoạn text bạn đưa cho AI, và nó phản hồi dựa trên đoạn text đó.

Hãy nghĩ về nó như **viết brief cho một copywriter freelance**:

- Brief mơ hồ ("viết gì đó về SEO") sẽ nhận lại nội dung chung chung, nhạt nhẽo
- Brief chi tiết (đối tượng, giọng văn, từ khóa, ví dụ) sẽ cho ra đúng thứ bạn cần

Nguyên tắc tương tự áp dụng cho AI. **Prompt càng tốt, output càng chất lượng.**

Trong dự án của chúng ta, prompt xuất hiện ở hai chỗ:

| Trong code Agno | Là gì | Ví dụ |
|---|---|---|
| `instructions=[...]` | **System prompt** -- hành vi cố định | "You are an expert SEO researcher" |
| `agent.run("...")` | **User prompt** -- yêu cầu cụ thể lượt này | "Research the topic: link building" |

Chúng ta sẽ phân tích cả hai trong bài học này.

## Giải phẫu một Prompt tốt

Một prompt tốt có **bốn thành phần**. Không phải lúc nào cũng cần đủ cả bốn, nhưng bạn đưa vào càng nhiều thì kết quả càng tốt:

1. **Role (Vai trò)** -- AI nên đóng vai ai
   - "You are an expert SEO writer with 10 years of experience"
   - Điều này thiết lập giọng điệu, từ vựng và mức độ chuyên môn

2. **Task (Nhiệm vụ)** -- Cần làm gì
   - "Write a meta description for an article about on-page SEO"
   - Càng cụ thể càng tốt: "Write 5 bullet points" hiệu quả hơn "Write some content"

3. **Constraints (Ràng buộc)** -- Các quy tắc cần tuân theo
   - "Maximum 160 characters", "Use active voice", "Do NOT use clickbait"
   - Ràng buộc giúp AI không đi lạc hướng

4. **Examples (Ví dụ)** -- Cho xem output tốt trông như thế nào (few-shot)
   - Đưa 1-3 ví dụ về chất lượng và định dạng bạn mong đợi
   - Kỹ thuật này gọi là **few-shot prompting** và hiệu quả đáng kinh ngạc

Hãy xem sự khác biệt mà nó tạo ra.

````python
# Prompt TỆ -- mơ hồ, không có cấu trúc
bad_prompt = "Write about SEO"

# Prompt TỐT -- có role, task, constraints, example
good_prompt = """You are an expert SEO content writer with 10 years of experience.

Write a meta description for an article about "on-page SEO optimization."

Constraints:
- Maximum 160 characters
- Include the keyword "on-page SEO" naturally
- Use active voice
- End with a call to action

Example of a good meta description:
"Master technical SEO with our step-by-step guide. Learn crawling, indexing, and site speed optimization. Start improving your rankings today."
"""

print(f"Prompt tệ:  {len(bad_prompt)} ký tự")
print(f"Prompt tốt: {len(good_prompt)} ký tự")
print(f"\nPrompt tốt dài gấp {len(good_prompt) // len(bad_prompt)} lần,")
print("nhưng sẽ cho kết quả TỐT HƠN RẤT NHIỀU!")
print()
print("--- Prompt tệ ---")
print(bad_prompt)
print()
print("--- Prompt tốt ---")
print(good_prompt)
````

## System Prompt vs User Prompt

Thực tế, LLM nhận **hai loại prompt** trong mỗi cuộc hội thoại:

**System prompt** = hành vi cố định. Đặt một lần. Tính cách và quy tắc của agent.
- Trong Agno: `instructions=["You are an SEO expert", "Answer concisely"]`
- Hãy xem nó như **bản mô tả công việc** -- luôn giữ nguyên cho mọi nhiệm vụ

**User prompt** = yêu cầu cụ thể lượt này. Thay đổi mỗi lần.
- Trong Agno: `agent.run("What is a meta description?")`
- Hãy xem nó như **yêu cầu công việc** -- khác nhau mỗi lần

Bên trong, LLM nhìn thấy cả hai kết hợp lại:

```
[System: You are an SEO expert. Answer concisely.]
[User: What is a meta description?]
   --> LLM xử lý cả hai cùng lúc --> Phản hồi
```

System prompt định hình **CÁCH** nó trả lời. User prompt xác định **NỘI DUNG** nó trả lời về cái gì.

````python
# Đây là những gì xảy ra bên trong khi agent chạy

system_prompt = "You are an SEO expert. Answer concisely and clearly."
user_prompt = "What is a meta description?"

# LLM nhìn thấy cả hai cùng lúc:
full_context = f"""--- System Prompt (từ instructions) ---
{system_prompt}

--- User Prompt (từ agent.run()) ---
{user_prompt}
"""

print(full_context)
print("LLM xử lý toàn bộ ngữ cảnh này để tạo ra phản hồi.")
print("System prompt định hình CÁCH nó trả lời.")
print("User prompt xác định NỘI DUNG nó trả lời về cái gì.")
````

## Prompt Engineering cho đội SEO

Dưới đây là năm kỹ thuật thực tế đặc biệt hiệu quả cho các tác vụ nội dung SEO:

### 1. Cụ thể luôn thắng dài dòng
"Write 5 bullet points about link building benefits" tốt hơn "Write some content about link building." Hãy chính xác về thứ bạn muốn.

### 2. Chỉ định định dạng output
Nói cho AI chính xác bạn muốn định dạng gì: "Return as JSON with keys: title, description, keywords" hoặc "Use H2 headings for each section." Nếu không chỉ định, bạn sẽ nhận được bất kỳ định dạng nào AI thích.

### 3. Ví dụ few-shot
Đưa 2-3 ví dụ về thứ bạn muốn. AI sẽ bắt chước theo pattern. Đây là một trong những kỹ thuật mạnh mẽ nhất và nó miễn phí -- bạn chỉ cần có ví dụ tốt.

### 4. Chỉ dẫn phủ định
"Do NOT use clickbait titles" hoặc "Do NOT start with 'In today's digital landscape'" hoạt động tốt đến bất ngờ. LLM rất giỏi trong việc tránh những thứ bạn yêu cầu rõ ràng.

### 5. Suy luận từng bước
"First analyze the keyword intent, then identify the target audience, then write the title" cho kết quả tốt hơn "Write a title." Chia nhỏ nhiệm vụ thành các bước sẽ dẫn dắt quá trình suy nghĩ của AI.

````python
# Đây là instructions THỰC TẾ từ Research Agent của chúng ta
# (từ output/agents/researcher.py)

research_instructions = [
    "You are an expert SEO researcher.",
    "Research the given topic using web search.",
    "Identify primary and secondary keywords, analyze what top-ranking content covers, "
    "and find content gaps.",
    "Return your findings as clear, organized research notes.",
]

print("Phân tích instructions của Research Agent:\n")
for i, instruction in enumerate(research_instructions, 1):
    print(f"  {i}. \"{instruction}\"")

print("\nChú ý pattern:")
print("  Dòng 1: ROLE       -- 'You are an expert SEO researcher'")
print("  Dòng 2: TASK       -- 'Research the given topic using web search'")
print("  Dòng 3-4: CONSTRAINTS -- bao gồm gì, định dạng ra sao")
````

## Prompt Template

Trong thực tế, bạn không viết prompt từ đầu mỗi lần. Thay vào đó, bạn xây dựng **prompt template** -- các hàm tái sử dụng nhận biến đầu vào và trả về chuỗi prompt đã được format.

Đây chính xác là cách các agent trong pipeline của chúng ta hoạt động. Phần `instructions` giữ nguyên, nhưng chủ đề, từ khóa và các chi tiết khác thay đổi theo mỗi lần chạy.

Prompt template chỉ đơn giản là một **hàm Python** sử dụng **f-string** (mà chúng ta đã học ở Bài học 1) để chèn biến vào một prompt viết sẵn.

````python
# Prompt template tái sử dụng dưới dạng hàm Python

def meta_description_prompt(topic, keyword, max_chars=160):
    """Tạo prompt để viết meta description."""
    return f"""You are an expert SEO copywriter.

Write a meta description for an article about "{topic}".

Requirements:
- Maximum {max_chars} characters
- Include the keyword "{keyword}" naturally
- Use active voice
- End with a call to action
- Make it compelling enough to click
"""

# Thử với các input khác nhau
prompt1 = meta_description_prompt("on-page SEO", "on-page SEO optimization")
prompt2 = meta_description_prompt("link building strategies", "link building")

print("=== Prompt 1 ===")
print(prompt1)
print("=== Prompt 2 ===")
print(prompt2)
print("Cùng template, khác input -- đây là cách agent hoạt động trong thực tế!")
````

## Prompt Engineering = Kỹ năng SEO mới

Nếu bạn là dân SEO, bạn đã sẵn có tư duy cho prompt engineering. Những điểm tương đồng rất rõ ràng:

| Viết SEO | Prompt Engineering |
|---|---|
| Bạn viết cho **thuật toán Google** | Bạn viết cho **LLM** |
| Bạn dùng **từ khóa** một cách chiến lược | Bạn dùng **instructions** một cách chiến lược |
| Bạn **test và lặp lại** với nội dung | Bạn **test và lặp lại** với prompt |
| Nội dung tốt hơn = thứ hạng cao hơn | Prompt tốt hơn = AI output tốt hơn |

Tóm lại:

**Prompt tốt hơn --> agent output tốt hơn --> ít chỉnh sửa thủ công hơn --> sản xuất nội dung quy mô lớn**

Đây là lý do prompt engineering quan trọng với đội SEO. Không phải để thay thế content writer -- mà để đưa cho AI bản brief tốt nhất có thể, để output cần chỉnh sửa tối thiểu.

## Tổng kết

Những gì bạn đã học trong bài học này:

- **Prompt** -- mọi thứ bạn gửi cho LLM (brief càng tốt, output càng chất lượng)
- **System prompt** (`instructions`) -- hành vi cố định, như bản mô tả công việc
- **User prompt** (`agent.run()`) -- yêu cầu cụ thể mỗi lần, như phiếu giao việc
- **4 thành phần** của một prompt tốt: **Role**, **Task**, **Constraints**, **Examples**
- **Prompt template** -- hàm Python tái sử dụng để tạo prompt bằng f-string
- **Prompt engineering cho SEO** -- cùng tư duy test-và-lặp-lại mà bạn đã quen thuộc

**Bài học tiếp theo:** Chúng ta sẽ tìm hiểu về các model AI khác nhau -- tại sao một số model phù hợp hơn cho những tác vụ nhất định, và cách chúng ta chọn model cụ thể cho từng bước trong pipeline.
