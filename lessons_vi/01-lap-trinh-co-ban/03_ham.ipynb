# Bài 3: Function — Tạo lệnh riêng của bạn

## Function là gì?

**Function** giống như một **công thức nấu ăn**. Bạn viết một lần, rồi tái sử dụng bao nhiêu lần tuỳ thích.

Ví dụ thực tế:
- **Công thức** "Pha cà phê" — làm theo mỗi khi bạn muốn uống cà phê
- **Function** `format_title()` — gọi mỗi khi bạn cần tạo tiêu đề SEO

Cú pháp cơ bản:
```python
def function_name(parameter):
    # code thực thi
    return result
```

- `def` = khai báo function
- `function_name` = tên do bạn đặt
- `parameter` = dữ liệu đầu vào
- `return` = trả về kết quả

````python
# Function đơn giản nhất — không có return
def greet(name):
    print(f"Hello {name}!")
    print(f"Welcome to the AI team, {name}.")

# Gọi function
greet("Viet")
print()
greet("Lan")
print()
greet("Minh")
````

## Argument và Return — Đầu vào và đầu ra

- **Argument** (parameter): Dữ liệu bạn truyền vào function
- **Return**: Kết quả mà function trả về

**Argument mặc định**: Bạn có thể đặt giá trị mặc định cho parameter. Nếu không truyền vào, Python sẽ dùng giá trị mặc định.

```python
def function(required, optional="default value"):
    return result
```

Trong dự án AI agent, nhiều function dùng argument mặc định, ví dụ `year=2026`, `status="draft"`,...

````python
def format_seo_title(topic, year=2026):
    """Create an SEO title from a topic and year."""
    return f"{topic} - Complete Guide {year}"

# Gọi với 1 argument (year mặc định là 2026)
title1 = format_seo_title("On-Page SEO")
print(title1)

# Gọi với 2 argument
title2 = format_seo_title("Content Marketing", 2025)
print(title2)

# Gọi với argument có tên (rõ ràng hơn)
title3 = format_seo_title(topic="Link Building", year=2026)
print(title3)
````

## Function với nhiều parameter

Function có thể nhận nhiều parameter. Đây là cách các agent trong dự án xử lý dữ liệu — nhận nhiều đầu vào, xử lý, rồi trả về kết quả.

**Mẹo hay:** `text.split()` tách string thành list các từ. `len()` đếm số phần tử. Kết hợp lại để đếm số từ!

````python
def word_count(text):
    """Count the number of words in a text."""
    words = text.split()
    return len(words)

# Thử với các đoạn văn bản khác nhau
sample1 = "SEO is the practice of optimizing a website to rank higher on Google"
sample2 = "Content marketing helps increase traffic organically"

print(f"Sample 1: {word_count(sample1)} words")
print(f"Sample 2: {word_count(sample2)} words")
print(f"Combined: {word_count(sample1) + word_count(sample2)} words")
````

## Tại sao function quan trọng?

1. **Tái sử dụng:** Viết một lần, dùng nhiều lần
2. **Tổ chức code:** Chia nhỏ thành từng phần dễ hiểu
3. **Bảo trì:** Sửa một chỗ, có hiệu lực ở mọi nơi

Trong dự án AI agent, mỗi agent thực chất là một function lớn:
- `content_writer` nhận chủ đề và trả về bài viết SEO hoàn chỉnh
- `image_finder` nhận bài viết và trả về bài viết đã bổ sung hình ảnh
- `aio_analyzer` nhận từ khóa và trả về gợi ý tối ưu cho AI Overview

Giờ hãy viết một function thực tế hơn!

````python
def create_article_summary(title, status, words):
    """Create a summary for an article."""
    status_icons = {
        "draft": "[DRAFT]",
        "review": "[REVIEW]",
        "published": "[DONE]",
        "error": "[ERROR]"
    }
    icon = status_icons.get(status, "[?]")
    
    summary = f"{icon} {title}\n"
    summary += f"    Status: {status}\n"
    summary += f"    Words: {words:,}\n"
    summary += f"    Read time: ~{words // 200} min"
    return summary

# Sử dụng function nhiều lần
print(create_article_summary("On-Page SEO 2026", "published", 2500))
print()
print(create_article_summary("Content Marketing Guide", "draft", 1800))
print()
print(create_article_summary("Technical SEO Basics", "review", 3200))
````
