# Bài 2: List và Dictionary

## List — Tập hợp có thứ tự

**List** giống như một **danh sách công việc**. Các phần tử được lưu theo thứ tự và đánh số từ `0`.

```python
my_list = ["item 1", "item 2", "item 3"]
#            [0]        [1]        [2]
```

**Lưu ý:** Python đếm từ `0`, không phải `1`!

Trong dự án AI agent, list dùng để lưu danh sách keyword, bài viết, các section, và nhiều thứ khác.

````python
keywords = ["SEO", "content marketing", "link building"]
print("All keywords:", keywords)
print("First keyword:", keywords[0])
print("Last keyword:", keywords[-1])
print("Total keywords:", len(keywords))
````

## Thêm và xoá phần tử trong list

- `list.append(x)` — Thêm `x` vào cuối list
- `list.remove(x)` — Xoá phần tử `x` khỏi list
- `list.insert(i, x)` — Chèn `x` vào vị trí `i`

List có thể thay đổi sau khi tạo — tính chất này gọi là **mutable** (có thể thay đổi).

````python
keywords = ["SEO", "content marketing", "link building"]
print("Original:", keywords)
keywords.append("on-page SEO")
keywords.append("technical SEO")
print("After adding:", keywords)
keywords.remove("link building")
print("After removing:", keywords)
keywords.insert(0, "Google ranking")
print("After inserting at front:", keywords)
````

## Dictionary (dict) — Cặp khoá-giá trị

**Dictionary** giống như một **tấm danh thiếp** hay **hồ sơ cá nhân**. Mỗi thông tin có một **tên** (key) và một **giá trị** (value).

```python
contact = {
    "name": "Viet",
    "age": 20,
    "company": "SEONGON"
}
```

Trong dự án, mỗi **bài viết** được lưu dưới dạng dict với các key như `title`, `status`, `word_count`,...

````python
article = {
    "title": "SEO Guide 2026",
    "status": "draft",
    "word_count": 1500,
    "author": "Viet"
}
print("Title:", article["title"])
print("Status:", article["status"])
print("Words:", article["word_count"])
article["status"] = "published"
print("New status:", article["status"])
article["views"] = 0
print("Full article:", article)
````

## Cấu trúc lồng nhau — Dữ liệu bên trong dữ liệu

Thực tế, dữ liệu thường phức tạp hơn nhiều. Dict có thể chứa list, list có thể chứa dict, cứ thế lồng vào nhau.

Ví dụ: một bài viết có nhiều mục, mỗi mục là một dict nằm trong một list.

Đó chính là cách dự án AI agent tổ chức dữ liệu bài viết!

Chúng ta sẽ xây dần theo hai bước:
1. Đầu tiên: **list chứa dict** (giống bảng tính)
2. Sau đó: **dict chứa list và dict** (pattern đầy đủ dùng trong dự án)

````python
# A list of dictionaries — like a spreadsheet where each row is a dict
articles = [
    {"title": "SEO Basics", "status": "published", "views": 1200},
    {"title": "Link Building", "status": "draft", "views": 0},
    {"title": "Content Strategy", "status": "published", "views": 800},
]

# Access a specific article (list index, then dict key)
print("First article:", articles[0]["title"])
print("Its status:", articles[0]["status"])
print(f"Total articles: {len(articles)}")
````

````python
# Now the full picture: a dict containing a list of dicts
# This is exactly how our project stores article data
article = {
    "title": "On-Page SEO Guide 2026",
    "author": "Viet",
    "keywords": ["on-page SEO", "optimization", "Google ranking"],
    "sections": [
        {"heading": "Title Tags", "word_count": 500},
        {"heading": "Meta Descriptions", "word_count": 400},
        {"heading": "Header Tags", "word_count": 600},
    ]
}

# Access nested data step by step
print(f"Article: {article['title']}")
print(f"First keyword: {article['keywords'][0]}")
print(f"First section: {article['sections'][0]['heading']}")
print(f"Its word count: {article['sections'][0]['word_count']}")
````

## Vòng lặp `for` — Duyệt qua từng phần tử

Vòng lặp **`for`** cho phép bạn thực hiện một thao tác với **từng phần tử** trong list.

```python
for item in my_list:
    # thực hiện thao tác với item
```

**Lưu ý:** Code bên trong vòng lặp phải được **thụt lề** (indent) 4 dấu cách. Đây là quy tắc bắt buộc của Python.

### `enumerate()` — Lấy số thứ tự

Khi bạn cần cả **số thứ tự** và **phần tử** trong vòng lặp, dùng `enumerate()`:

```python
for i, keyword in enumerate(keywords, 1):
```

Mỗi vòng lặp cho bạn hai giá trị:
- `i` — số thứ tự (bắt đầu từ 1, vì ta truyền `1`)
- `keyword` — phần tử đó

Không có `enumerate`, bạn chỉ nhận được phần tử và phải tự đếm số thứ tự.

````python
keywords = ["on-page SEO", "content marketing", "link building", "technical SEO", "local SEO"]
print("=== Keyword List ===")
print()
for i, keyword in enumerate(keywords, 1):
    print(f"  {i}. {keyword}")
print()
print(f"Total: {len(keywords)} keywords")
print()
print("SEO-related keywords:")
for keyword in keywords:
    if "SEO" in keyword:
        print(f"  -> {keyword}")
````
