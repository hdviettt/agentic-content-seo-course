## Chỗ no-code không đi tiếp được

Bạn hoàn toàn xây được một content pipeline cơ bản trên n8n. AI node → Airtable node → trigger → xong. Tự động hóa đơn giản? n8n nhanh hơn code. Không bàn cãi.

Nhưng đây là chỗ nó vỡ:

**Tốc độ batch.** n8n chạy từng item một. Không song song được — đó là giới hạn nền tảng, không phải bug. 20 bài × 5 phút = 100 phút chờ. Với code, chạy 3-5 bài cùng lúc. Cùng batch, ~30 phút.

**Structured output.** Output parser của n8n không đáng tin. JSON sai cấu trúc, lỗi validation sau vài lượt, hỏng khi gặp Markdown trong chuỗi. Issue trên GitHub, có cái mở hơn 2 năm. Với code, bạn buộc AI xuất đúng cấu trúc ngay khi tạo. Không phải "thường đúng." Luôn đúng.

**Hội thoại.** Trong n8n, mỗi lần chạy workflow bắt đầu từ số 0. Bạn không nói được "bài đó lỗi, thử lại đi" — không có "bài đó." Bạn phải mở dashboard, tìm execution #47, bấm retry. Với code, bạn xây hệ thống nhớ ngữ cảnh. Bạn nói chuyện với nó. Nó hành động.

Đây không phải thứ n8n sẽ fix trong bản cập nhật. Đây là giới hạn của cách workflow builder trực quan hoạt động.

n8n đưa bạn đến đây. Code đưa bạn đi tiếp.

## Variable — Những chiếc hộp có nhãn

**Variable** giống như một **chiếc hộp có dán nhãn**. Bạn đặt tên cho hộp, rồi bỏ dữ liệu vào bên trong.

Ví dụ:
- Hộp dán nhãn `name` chứa giá trị `"Viet"`
- Hộp dán nhãn `age` chứa giá trị `20`

Dấu `=` có nghĩa là “gán giá trị” — không giống dấu “bằng” trong toán học.

````python
name = "Viet"
print(name)

age = 20
print(age)
````

## String và f-string — Chèn biến vào văn bản

**String** là dữ liệu dạng văn bản, luôn được bao bởi dấu ngoặc kép: `"..."` hoặc `'...'`.

**f-string** cho phép bạn chèn giá trị của variable vào trong string. Chỉ cần thêm chữ `f` trước dấu ngoặc mở và đặt tên variable trong `{}`.

```python
f"Hello {variable_name}"
```

Cách viết này dùng rất nhiều trong dự án của chúng ta.

````python
company = "SEONGON"
role = "AI Leader"

print(f"{name} works at {company}")
print(f"Role: {role}")
print(f"{name} - {role} @ {company}")
````

## Số — Phép tính cơ bản

Python có thể tính toán như máy tính bỏ túi:

| Phép tính    | Ký hiệu | Ví dụ          |
|-------------|--------|----------------|
| Cộng        | `+`    | `5 + 3` = 8    |
| Trừ        | `-`    | `10 - 4` = 6   |
| Nhân        | `*`    | `3 * 7` = 21   |
| Chia        | `/`    | `20 / 4` = 5.0 |

Bạn có thể kết hợp số với f-string để hiển thị kết quả dễ đọc hơn.

````python
articles = 5
words_per_article = 2000
total_words = articles * words_per_article

print(f"Articles: {articles}")
print(f"Words per article: {words_per_article}")
print(f"Total words: {total_words}")
print(f"If you write {articles} articles/week, that's {articles * 4} per month!")
````
