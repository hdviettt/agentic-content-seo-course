# Bài học 9: Agent với Tools

## Tại Sao Agent Cần Tools?

Ở bài trước, agent của chúng ta chỉ có thể **suy nghĩ** và **trả lời** dựa trên kiến thức sẵn có.

Nhưng nếu bạn hỏi: *"Xu hướng SEO mới nhất năm 2026 là gì?"* — Agent không biết vì kiến thức của nó có giới hạn thời gian!

Ở Bài học 5, ta đã biết rằng LLM được huấn luyện trên dữ liệu đến một ngày nhất định — sau đó, chúng hoàn toàn mù thông tin. **Tools** chính là cách ta giải quyết vấn đề này. Chúng cho phép agent **làm** được nhiều hơn:
- Tìm kiếm trên web
- Truy vấn cơ sở dữ liệu
- Tìm hình ảnh
- Và nhiều thứ khác...

Giống như một nhân viên có thể **đọc tin tức và tìm kiếm trên Google**, agent có tools cũng có thể tìm kiếm thông tin theo thời gian thực.

> **Chi phí:** ~$0.02-0.10 mỗi ô (Sonnet gọi với web search). Chạy mỗi ô một lần để giữ chi phí thấp.

## DuckDuckGo — Công Cụ Tìm Kiếm Miễn Phí

**DuckDuckGo** là một công cụ tìm kiếm web:
- Hoàn toàn **miễn phí** — không cần API key
- Agent **tự động tìm kiếm** khi cần thông tin mới
- Kết quả được agent **đọc hiểu và tóm tắt** cho bạn

Thêm tool vào agent rất đơn giản:
```python
agent = Agent(
    tools=[DuckDuckGoTools()],  # Chỉ cần thêm dòng này!
)
```

````python
from dotenv import load_dotenv
load_dotenv()

from agno.agent import Agent
from agno.models.anthropic import Claude
from agno.tools.duckduckgo import DuckDuckGoTools

# Agent có khả năng tìm kiếm web!
agent = Agent(
    name="Researcher",
    model=Claude(id="claude-sonnet-4-5-20250929"),
    tools=[DuckDuckGoTools()],
    instructions=["Search the web for information and summarize the results."],
)

response = agent.run("Find the top 3 SEO trends in 2026")
print(response.content)
````

## Chuyện Gì Vừa Xảy Ra?

Khi bạn chạy đoạn code ở trên, đây là những gì diễn ra phía sau:

1. **Bạn gửi một câu hỏi** — "Find the top 3 SEO trends in 2026"
2. **Agent suy nghĩ** — "Mình cần thông tin mới, dùng DuckDuckGo thôi"
3. **Agent tự động gọi tool** — Tìm kiếm trên DuckDuckGo
4. **Agent đọc kết quả** — Xử lý các trang web tìm được
5. **Agent tóm tắt** — Viết câu trả lời dựa trên những gì tìm được

**Điểm mấu chốt:** Bạn KHÔNG cần tự viết code tìm kiếm. Agent **tự động quyết định** khi nào cần dùng tools!

````python
# Ví dụ khác — nghiên cứu một chủ đề cụ thể
response = agent.run("What is SEONGON? Find information about this company.")
print(response.content)
````

## So Sánh: Agent KHÔNG CÓ Tools vs CÓ Tools

Hãy xem sự khác biệt khi agent có và không có tools:

| | Không có Tools | Có Tools |
|---|---|---|
| Kiến thức | Chỉ biết những gì được huấn luyện | Có thể tìm kiếm thông tin mới |
| Thông tin | Có thể lỗi thời/sai | Cập nhật và chính xác |
| Khả năng | Chỉ suy nghĩ | Suy nghĩ + Hành động |

Chạy code bên dưới để thấy sự khác biệt rõ ràng!

````python
# Agent KHÔNG CÓ tools
agent_no_tools = Agent(
    name="Agent Without Tools",
    model=Claude(id="claude-sonnet-4-5-20250929"),
    instructions=["Answer concisely."],
)

# Agent CÓ tools
agent_with_tools = Agent(
    name="Agent With Tools",
    model=Claude(id="claude-sonnet-4-5-20250929"),
    tools=[DuckDuckGoTools()],
    instructions=["Search the web then answer concisely."],
)

question = "Which website currently ranks #1 for the keyword 'learn SEO' on Google?"

print("=== AGENT KHÔNG CÓ TOOLS ===")
r1 = agent_no_tools.run(question)
print(r1.content)

print("\n=== AGENT CÓ TOOLS ===")
r2 = agent_with_tools.run(question)
print(r2.content)
````

## Tổng Kết Bài học 9

Những gì bạn đã học:
- **Tools** giúp agent **hành động**, không chỉ suy nghĩ
- **DuckDuckGoTools** — tìm kiếm web miễn phí, không cần API key
- Agent **tự động quyết định** khi nào cần dùng tools
- Agent có tools cho kết quả **chính xác và cập nhật hơn**

**Bài tiếp theo:** Chúng ta sẽ học cách khiến agent trả về **dữ liệu có cấu trúc** (structured output) — thay vì văn bản tự do!
