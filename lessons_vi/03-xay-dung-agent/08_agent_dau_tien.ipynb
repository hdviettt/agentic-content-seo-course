Cứ nghĩ thế này: agency của bạn phụ trách hàng chục khách hàng, mỗi khách cần nội dung thường xuyên. Một bài viết tốn 4-8 tiếng nghiên cứu, viết, và tối ưu hóa. ## So Sánh Các Cách Tiếp Cận

| Cách tiếp cận | Tốc độ | Chất lượng | Khả năng mở rộng | Hạn chế |
|----------------|--------|------------|-------------------|---------|
| Viết tay | 4-8 tiếng | Cao (bạn kiểm soát) | Không mở rộng được | Thời gian |

Điểm mấu chốt: mọi cách tự động hóa khác dùng **một lần gọi AI** để làm tất cả. Pipeline của chúng ta dùng **agent chuyên biệt**, mỗi agent làm tốt nhất việc của mình — giống như một team nội dung thực tế có người nghiên cứu, người viết bài, và người biên tập.

Cột "Hạn chế" là phần quan trọng: **cách tiếp cận này đòi hỏi hiểu biết để xây dựng và tùy chỉnh**. Đó chính là điều khóa học này dạy bạn.

# Bài học 8: Agent AI Đầu Tiên Của Bạn

## Agent AI là gì?

Một **agent AI** là một chương trình có khả năng **suy nghĩ** và **phản hồi** — giống như một trợ lý thông minh.

Nhớ lại Bài học 5: một LLM dự đoán token tiếp theo dựa trên mọi thứ nó đã thấy. Khi bạn tạo một agent, bạn đang gán cho LLM đó một **vai trò** và **chỉ dẫn** để điều hướng output của nó. Đó là điều xảy ra bên dưới mỗi lần agent chạy.

Hãy tưởng tượng việc tuyển một nhân viên mới:
- Bạn **đặt tên cho họ** (name)
- Bạn **chọn bộ não cho họ** (model — ở đây là Claude)
- Bạn **viết mô tả công việc** (instructions — system prompt từ Bài học 6)
- Rồi bạn **giao việc** và họ bắt tay vào làm (run)

Đó chính là cách agent AI hoạt động!

Trong bài học này, chúng ta sẽ:
1. Tạo agent AI đầu tiên
2. Gửi tin nhắn cho nó
3. Thay đổi hành vi của agent bằng `instructions`

## Cấu Trúc Agent Cơ Bản

Một agent trong Agno có 3 thành phần chính:

```python
Agent(
    name="...",          # Tên của agent
    model=Claude(...),    # Bộ não — model AI nào sẽ thực hiện suy luận
    instructions=[...],   # Mô tả công việc — agent nên hành xử ra sao
)
```

- **`name`**: Định danh cho agent. Giống như tên nhân viên.
- **`model`**: Model AI nào thực hiện suy luận. Chúng ta dùng Claude Sonnet.
- **`instructions`**: Danh sách các chỉ dẫn. Giống như viết mô tả công việc cho nhân viên mới.

> **Lưu ý về chi phí:** Từ bài học này trở đi, các cell gọi `agent.run()` sẽ tạo API call tốn tiền (~$0.01-0.05 mỗi lần gọi với Sonnet). Chạy mỗi cell một lần để xem kết quả — tránh chạy lại không cần thiết.

````python
from dotenv import load_dotenv
load_dotenv()

from agno.agent import Agent
from agno.models.anthropic import Claude

# Tạo agent đầu tiên!
agent = Agent(
    name="SEO Assistant",
    model=Claude(id="claude-sonnet-4-5-20250929"),
    instructions=["You are an SEO expert. Answer concisely and clearly."],
)

# Gửi tin nhắn cho agent
response = agent.run("What is a meta description? Explain in 2 sentences.")
print(response.content)
````

## Phân Tích Code

Hãy xem đoạn code ở trên từng phần:

| Code | Ý nghĩa |
|------|----------|
| `name="SEO Assistant"` | Đặt tên agent là "SEO Assistant" |
| `model=Claude(id="claude-sonnet-4-5-20250929")` | Dùng Claude Sonnet — nhanh và thông minh |
| `instructions=["..."]` | Chỉ dẫn: "hãy là chuyên gia SEO, trả lời ngắn gọn" |
| `agent.run("...")` | Gửi câu hỏi và nhận phản hồi |
| `response.content` | Nội dung câu trả lời của agent |

**Lưu ý:** `instructions` là một **list** — bạn có thể truyền nhiều chỉ dẫn.

````python
# Tạo agent khác với instructions khác
content_agent = Agent(
    name="Content Assistant",
    model=Claude(id="claude-sonnet-4-5-20250929"),
    instructions=[
        "You are a content marketing expert.",
        "Answer clearly, suitable for beginners.",
    ],
)

response = content_agent.run("How do you write a compelling article title?")
print(response.content)
````

## Instructions — "Mô Tả Công Việc" Của Agent

`instructions` là phần **quan trọng nhất** của agent. Nó giống như **mô tả công việc (JD)** khi tuyển dụng.

**JD tốt** = nhân viên làm đúng việc. **Instructions tốt** = agent phản hồi đúng cách.

Ví dụ:
- `"Answer concisely"` — Agent trả lời ngắn gọn
- `"Always respond using bullet points"` — Agent dùng bullet points
- `"You are an SEO expert with 10 years of experience"` — Agent cho kiến thức SEO chuyên sâu

Hãy thử thay đổi instructions để xem sự khác biệt!

````python
# Agent với instructions cụ thể: LUÔN dùng bullet points
agent_bullet = Agent(
    name="Bullet Point Expert",
    model=Claude(id="claude-sonnet-4-5-20250929"),
    instructions=[
        "ALWAYS respond using bullet points.",
        "Each bullet point should be at most 1 sentence.",
    ],
)

# Agent với instructions khác: giải thích như đang nói chuyện với học sinh
agent_simple = Agent(
    name="Simple Explainer",
    model=Claude(id="claude-sonnet-4-5-20250929"),
    instructions=[
        "Explain concepts as if talking to a middle school student.",
        "Use real-world examples that are easy to understand.",
    ],
)

question = "What are backlinks and why are they important for SEO?"

print("=== AGENT BULLET POINT ===")
r1 = agent_bullet.run(question)
print(r1.content)

print("\n=== AGENT GIẢI THÍCH ĐƠN GIẢN ===")
r2 = agent_simple.run(question)
print(r2.content)
````
