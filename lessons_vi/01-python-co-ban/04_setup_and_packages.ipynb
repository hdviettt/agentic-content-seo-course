## Package là gì?

**Package** (thư viện) là những đoạn code được viết sẵn mà bạn có thể **cài đặt và sử dụng**.

Hãy hình dung như việc mua nguyên liệu chế biến sẵn thay vì tự làm mọi thứ từ đầu:
- Thay vì tự viết code gọi API, dùng package `openai`
- Thay vì tự viết code đọc file .env, dùng `python-dotenv`
- Thay vì tự xây framework cho AI agent, dùng `agno`

**pip** là trình cài đặt package cho Python (giống như một kho ứng dụng dành riêng cho Python).

**Lưu ý:** Trên máy của chúng ta, hãy dùng `python -m pip` thay vì chỉ `pip` (vì pip không nằm trên PATH).

## Import — Đưa Package vào code của bạn

Sau khi cài đặt, bạn cần **import** package để sử dụng.

Có một số cách import:

```python
import os                          # Import toàn bộ package
from dotenv import load_dotenv     # Import một hàm cụ thể
import json as j                   # Import với tên viết tắt
```

**`os`** là một package có sẵn trong Python, dùng để tương tác với hệ điều hành: đọc file, lấy đường dẫn, đọc biến môi trường, v.v.

````python
import os

# Lấy thư mục hiện tại
current_dir = os.getcwd()
print(f"Current directory: {current_dir}")

# Liệt kê các file trong thư mục
files = os.listdir(current_dir)
print(f"\nFiles/folders:")
for f in files[:10]:  # Chỉ hiển thị 10 mục đầu tiên
    print(f"  - {f}")

# Kiểm tra hệ điều hành
print(f"\nOperating system: {os.name}")
````

## File .env — Lưu trữ bí mật một cách an toàn

**API key** là "mật khẩu" để sử dụng các dịch vụ AI (Claude, Grok, Freepik, v.v.).

**TUYỆT ĐỐI KHÔNG** đặt API key trực tiếp trong code! Tại sao?
- Nếu bạn đẩy code lên GitHub, ai cũng có thể thấy key của bạn
- Người khác có thể dùng key của bạn và bạn phải trả tiền

**Giải pháp:** Lưu key trong file `.env` (file biến môi trường).
- File `.env` chỉ nằm trên máy của bạn
- File `.gitignore` ngăn không cho `.env` bị đẩy lên GitHub
- Code đọc key từ `.env` thông qua package `python-dotenv`

````python
# Đây là nội dung file .env trong dự án của chúng ta
# (ĐỪNG chạy ô này — đây chỉ là ví dụ minh họa)

env_example = """
# === .env file ===
# API key cho các dịch vụ AI

ANTHROPIC_API_KEY=sk-ant-api03-xxxxxxxxxxxx
XAI_API_KEY=xai-xxxxxxxxxxxx

# Tùy chọn — dành cho hình ảnh
FREEPIK_API_KEY=fpk-xxxxxxxxxxxx
DATA_FOR_SEO_API_KEY=Basic dXNlcjpwYXNz
"""

print(env_example)
print("Lưu ý: Các giá trị 'xxx' ở trên chỉ là ví dụ.")
print("Key thật nằm trong file .env của dự án (không bao giờ được commit lên git).")
````

## Tải .env bằng python-dotenv

Package `python-dotenv` đọc file `.env` và nạp các giá trị vào **biến môi trường**.

Sau đó bạn dùng `os.getenv("KEY_NAME")` để lấy giá trị.

```python
os.getenv("KEY")              # Trả về None nếu không tìm thấy
os.getenv("KEY", "default")   # Trả về "default" nếu không tìm thấy
```

**Mẹo bảo mật:** Khi in API key ra màn hình, hãy luôn **che bớt** để không lộ toàn bộ key.

````python
from dotenv import load_dotenv
import os

# Nạp file .env từ thư mục dự án
load_dotenv()

# Đọc API key (có giá trị dự phòng nếu chưa thiết lập)
api_key = os.getenv("ANTHROPIC_API_KEY", "not-set")

# Che bớt key khi in (chỉ hiển thị 8 ký tự đầu)
if len(api_key) > 8:
    masked = api_key[:8] + "..."
else:
    masked = api_key

print(f"API Key: {masked}")

# Kiểm tra tất cả các key
keys_to_check = ["ANTHROPIC_API_KEY", "XAI_API_KEY", "FREEPIK_API_KEY", "DATA_FOR_SEO_API_KEY"]

print("\nTrạng thái API key:")
for key_name in keys_to_check:
    value = os.getenv(key_name)
    if value:
        print(f"  {key_name}: Đã thiết lập")
    else:
        print(f"  {key_name}: Chưa thiết lập")
````

## Kiểm tra thiết lập

Trước khi chuyển sang Module 2, hãy chạy ô bên dưới để xác nhận mọi thứ hoạt động đúng. **Cả 4 mục đều phải đạt.** Nếu có mục nào không đạt, hãy sửa trước khi tiếp tục — các module sau sẽ không chạy được nếu thiếu thiết lập đầy đủ.

````python
from dotenv import load_dotenv
import os

load_dotenv()

checks_passed = 0
total_checks = 4

# Kiểm tra 1: Các package cốt lõi
print("1. Các package cốt lõi...")
try:
    import agno, anthropic, pydantic, ddgs, httpx
    print("   ĐẠT - Tất cả package cốt lõi đã được cài đặt")
    checks_passed += 1
except ImportError as e:
    print(f"   CHƯA ĐẠT - Thiếu package: {e.name}")
    print(f"   Cách sửa: python -m pip install -r requirements.txt")

# Kiểm tra 2: API key của Anthropic
print("2. Anthropic API key...")
key = os.getenv("ANTHROPIC_API_KEY", "")
if len(key) > 10:
    print(f"   ĐẠT - Đã tìm thấy key ({key[:8]}...)")
    checks_passed += 1
else:
    print("   CHƯA ĐẠT - Không tìm thấy ANTHROPIC_API_KEY trong .env")
    print("   Cách sửa: Thêm key vào file .env ở thư mục gốc của dự án")

# Kiểm tra 3: API key của xAI
print("3. xAI API key...")
key = os.getenv("XAI_API_KEY", "")
if len(key) > 5:
    print(f"   ĐẠT - Đã tìm thấy key ({key[:8]}...)")
    checks_passed += 1
else:
    print("   CHƯA ĐẠT - Không tìm thấy XAI_API_KEY trong .env")
    print("   Cách sửa: Thêm key vào file .env ở thư mục gốc của dự án")

# Kiểm tra 4: Package openai (cần cho xAI provider)
print("4. Package OpenAI (cần cho Grok)...")
try:
    import openai
    print("   ĐẠT - Package openai đã được cài đặt")
    checks_passed += 1
except ImportError:
    print("   CHƯA ĐẠT - Package openai chưa được cài đặt")
    print("   Cách sửa: python -m pip install openai")

print()
print(f"Kết quả: {checks_passed}/{total_checks} mục đạt")
if checks_passed == total_checks:
    print("Bạn đã sẵn sàng cho Module 2!")
else:
    print("Hãy sửa các mục chưa đạt trước khi tiếp tục.")
````
