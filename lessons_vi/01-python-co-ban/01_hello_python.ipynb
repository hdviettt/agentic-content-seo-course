## Variable — Những chiếc hộp có nhãn

**Variable** giống như một **chiếc hộp có dán nhãn**. Bạn đặt tên cho hộp, rồi bỏ dữ liệu vào bên trong.

Ví dụ:
- Hộp dán nhãn `name` chứa giá trị `"Viet"`
- Hộp dán nhãn `age` chứa giá trị `20`

Dấu `=` có nghĩa là “gán giá trị” — không giống dấu “bằng” trong toán học.

````python
name = "Viet"
print(name)

age = 20
print(age)
````

## String và f-string — Chèn biến vào văn bản

**String** là dữ liệu dạng văn bản, luôn được bao bởi dấu ngoặc kép: `"..."` hoặc `'...'`.

**f-string** cho phép bạn chèn giá trị của variable vào trong string. Chỉ cần thêm chữ `f` trước dấu ngoặc mở và đặt tên variable trong `{}`.

```python
f"Hello {variable_name}"
```

Cách viết này được dùng rất nhiều trong toàn bộ mã nguồn dự án của chúng ta.

````python
company = "SEONGON"
role = "AI Leader"

print(f"{name} works at {company}")
print(f"Role: {role}")
print(f"{name} - {role} @ {company}")
````

## Số — Phép tính cơ bản

Python có thể tính toán như một chiếc máy tính bỏ túi:

| Phép tính    | Ký hiệu | Ví dụ          |
|-------------|--------|----------------|
| Cộng        | `+`    | `5 + 3` = 8    |
| Trừ        | `-`    | `10 - 4` = 6   |
| Nhân        | `*`    | `3 * 7` = 21   |
| Chia        | `/`    | `20 / 4` = 5.0 |

Bạn có thể kết hợp số với f-string để hiển thị kết quả dễ đọc hơn.

````python
articles = 5
words_per_article = 2000
total_words = articles * words_per_article

print(f"Articles: {articles}")
print(f"Words per article: {words_per_article}")
print(f"Total words: {total_words}")
print(f"If you write {articles} articles/week, that's {articles * 4} per month!")
````
