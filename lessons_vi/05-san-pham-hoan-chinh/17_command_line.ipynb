## CLI là gì?

CLI (Command Line Interface) là cách chạy chương trình từ **terminal** mà không cần viết code.

Thay vì mở Jupyter notebook, viết code rồi chạy từng cell, bạn chỉ cần gõ **một lệnh duy nhất**:

```
python output/cli.py create "SEO Guide 2026"
```

CLI phù hợp khi:
- Bạn đã biết mình muốn làm gì (tạo bài viết, kiểm tra trạng thái...)
- Bạn muốn **tự động hóa** tác vụ (ví dụ: lên lịch chạy hàng ngày)
- Bạn muốn chạy nhanh mà không cần mở trình soạn code

File `cli.py` là một **lớp bọc mỏng** (thin wrapper) — nó chỉ nhận lệnh rồi gọi `pipeline.py` và `db.py` để thực hiện công việc thật sự.

## Bên Trong Hoạt Động Như Thế Nào

CLI sử dụng package `argparse` có sẵn của Python để phân tích lệnh. Đây là pattern cơ bản:

```python
# 1. Định nghĩa các lệnh có thể dùng
parser = argparse.ArgumentParser()
subparsers = parser.add_subparsers()

# 2. Thêm một lệnh với các tùy chọn
p = subparsers.add_parser("status")
p.add_argument("--filter", help="Lọc theo trạng thái")

# 3. Phân tích nội dung người dùng nhập và gọi hàm tương ứng
args = parser.parse_args()
args.func(args)
```

Mỗi lệnh CLI là một **hàm đơn giản** gọi lại code bạn đã biết:
- `create` gọi `create_article()` từ `db.py`, sau đó `run_content_pipeline()` từ `pipeline.py`
- `status` gọi `list_articles()` hoặc `get_article()` từ `db.py`
- `history` gọi `get_article_versions()` từ `db.py`

CLI **không chứa bất kỳ logic nghiệp vụ nào** — nó chỉ là giao diện thân thiện để gọi các hàm bạn đã học.

Dưới đây là các cách sử dụng — chỉ là ví dụ minh họa, **không chạy thật** để tiết kiệm chi phí API:

## Tổng Kết

- **CLI** (`cli.py`) chạy pipeline từ terminal — nhanh và dễ tự động hóa
- CLI là **lớp bọc mỏng** — không chứa logic nghiệp vụ
- Mỗi lệnh CLI chỉ gọi hàm từ `pipeline.py` (tạo bài viết) hoặc `db.py` (kiểm tra trạng thái, lịch sử)
- Các lệnh chính: `create`, `create-batch`, `status`, `history`, `chat`
- Tạo nhanh một bài: dùng `create`
- Tạo nhiều bài: dùng `create-batch`
- Kiểm tra kết quả: dùng `status`

**Bài học tiếp theo:** Chúng ta sẽ khám phá **giao diện chat** — một đội AI mà bạn có thể trò chuyện tự nhiên!
