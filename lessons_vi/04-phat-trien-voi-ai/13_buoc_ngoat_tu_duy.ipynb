Câu trả lời: **AI-assisted development** (phát triển với sự hỗ trợ của AI).

Bài học này giải thích bước chuyển đổi đó. Không có code để chạy, không cần API key — chỉ là một cách tư duy mới về cách làm việc với AI.

## Cách bạn đang dùng AI

Bạn đã dùng AI hàng ngày. Nghĩ về quy trình hiện tại:

- **Ý tưởng nội dung** — "Cho tôi 10 chủ đề blog về technical SEO"
- **Nghiên cứu** — "Tóm tắt chiến lược nội dung của đối thủ này"
- **Viết lại** — "Làm đoạn văn này hấp dẫn hơn"
- **Brainstorming** — "Tôi nên target keyword nào cho [chủ đề]?"

Quy trình trông thế này:

1. Mở trình duyệt (ChatGPT, Gemini, v.v.)
2. Dán text hoặc gõ câu hỏi
3. Nhận kết quả
4. Copy phần hữu ích vào tài liệu

Đây là **conversation-based AI** (AI dạng hội thoại) — bạn hỏi, nó trả lời, bạn copy phần cần dùng.

Cách này hiệu quả cho tác vụ đơn lẻ. Nhưng khi cần sửa 5 file trong codebase thì sao? Copy-paste bắt đầu bất lực. Bạn trở thành "người trung gian" — copy code từ AI, dán vào đúng file, hy vọng không sót gì, rồi quay lại lấy file tiếp theo.

Có cách tốt hơn.

## Công cụ AI cho phát triển phần mềm — toàn cảnh

Có nhiều loại công cụ AI hỗ trợ viết code. Mỗi loại hoạt động khác nhau:

| Công cụ | Cách hoạt động | Phù hợp cho | Hạn chế |
|---------|---------------|-------------|--------|
| **ChatGPT / Gemini** | Hội thoại trên trình duyệt | Câu hỏi đơn lẻ, nội dung, brainstorming | Không thấy và không sửa được file |
| **GitHub Copilot** | Gợi ý tự động trong editor | Viết code từng dòng | Chỉ thấy file đang mở |
| **Cursor** | Code editor tích hợp AI | Chỉnh sửa code với gợi ý AI | Cần học cách dùng editor |
| **Claude Code** | AI agent trong terminal | Thay đổi toàn dự án, sửa nhiều file | Dòng lệnh (không có GUI) |

### Mỗi công cụ trông như thế nào trong thực tế

**ChatGPT / Gemini:**
> Bạn gõ: "Viết hàm Python tính thời gian đọc từ số từ"
> Nó trả về code. Bạn copy thủ công vào file.

**GitHub Copilot:**
> Bạn bắt đầu gõ `def calculate_reading_time` trong editor. Nó gợi ý phần còn lại bằng chữ xám. Bạn nhấn Tab để chấp nhận.

**Cursor:**
> Bạn chọn một đoạn code, nhấn phím tắt, gõ "refactor thành class." Nó hiển thị diff — những gì muốn thay đổi. Bạn chấp nhận hoặc từ chối.

**Claude Code:**
> Bạn gõ: "Thêm trường reading_time vào article storage và cập nhật tất cả agent liên quan." Nó đọc `storage.py`, `content_writer.py`, `team.py`, lên kế hoạch, rồi sửa cả 3 file. Bạn review và duyệt.

Hãy chú ý sự khác biệt về **phạm vi**: ChatGPT không thấy gì. Copilot thấy một file. Cursor thấy file bạn chọn. Claude Code thấy **toàn bộ dự án**.

## Tại sao dùng Claude Code trong khóa học này

Khóa học này dùng Claude Code vì ba lý do cụ thể:

### 1. Nắm bắt toàn bộ dự án

Claude Code đọc toàn bộ codebase. Nó hiểu cách các file kết nối với nhau. Khi bạn nói "thêm tool mới," nó biết cần tạo file nào, import từ đâu, và tuân theo pattern nào — vì nó thấy tất cả.

### 2. Bạn chỉ đạo, nó xây dựng

Bạn dùng kiến thức từ Mô-đun 1-3 (Python, khái niệm AI, agent, tool, lưu trữ) để đưa ra chỉ dẫn thông minh. Claude Code lo phần triển khai. Bạn không cần gõ từng dòng code — bạn cần biết **yêu cầu gì**.

### 3. Không cần học editor mới

Nếu bạn đã dùng ChatGPT, bạn biết cách gõ chỉ dẫn và đọc kết quả. Claude Code hoạt động y vậy — bạn gõ yêu cầu, nó thực hiện. Điểm khác duy nhất: nó chạy trong terminal thay vì trình duyệt, và nó thực sự sửa được file.

> **Lưu ý**: Claude Code không phải lựa chọn duy nhất. Khóa học này dùng nó vì phù hợp với cách tiếp cận giảng dạy — bạn học khái niệm, rồi chỉ đạo AI triển khai. Nếu sau này bạn thích Cursor hay Copilot hơn, kiến thức vẫn áp dụng được. Kỹ năng **chỉ đạo AI hiệu quả** dùng được với mọi công cụ.

```python
# So sánh bài tập notebook và sản phẩm thật

notebook_exercise = {
    "files": 1,  # Một ô notebook
    "lines_of_code": 10,
    "concepts": ["one agent", "one tool"],
}

real_product = {
    "files": 17,  # output/backend/ + output/frontend/
    "lines_of_code": "~2,000",
    "concepts": ["3 agents", "4 tool modules", "web server", "React frontend", "storage layer"],
}

print("Notebook exercise:")
for k, v in notebook_exercise.items():
    print(f"  {k}: {v}")

print("\nReal product (output/):")
for k, v in real_product.items():
    print(f"  {k}: {v}")

print("\nThis is why you need AI-assisted development tools.")
print("You understand the concepts. The tool handles the complexity.")
```

## Kiến trúc sư và đội thi công

Đây là mô hình tư duy quan trọng nhất cho toàn bộ mô-đun:

- **Bạn là kiến trúc sư** — bạn hiểu hệ thống cần làm gì, agent hoạt động ra sao, cần tool nào, dữ liệu chảy thế nào
- **Claude Code là đội thi công** — nó viết code, tạo file, chạy test

Kiến trúc sư không xây gạch. Nhưng kiến trúc sư không hiểu vật liệu xây dựng thì vẽ bản thiết kế tệ.

Đó là lý do Mô-đun 1-3 quan trọng — chúng dạy bạn "vật liệu":

- **Mô-đun 1** (Python cơ bản) — Bạn đọc được bản vẽ (biến, hàm, dictionary, package)
- **Mô-đun 2** (Nền tảng AI) — Bạn hiểu vật liệu (LLM, token, prompt, model)
- **Mô-đun 3** (Xây dựng agent) — Bạn hiểu các thành phần (agent, tool, structured output, chuỗi xử lý, lưu trữ)

Từ giờ, vai trò của bạn thay đổi:

| Trước (Mô-đun 1-3) | Sau (Mô-đun 4+) |
|---------------------|------------------|
| "Viết code Python này" (bạn code trực tiếp) | "Yêu cầu Claude Code xây dựng" (bạn chỉ đạo, nó code) |

Giá trị của bạn nằm ở **biết cần yêu cầu gì**, không phải ở việc tự gõ code.
