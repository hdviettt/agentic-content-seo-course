## Claude Code là gì?

Claude Code là **trợ lý AI dạng command-line** do Anthropic phát triển. Điểm khác biệt so với ChatGPT:

| Tính năng | ChatGPT / Web Chat | Claude Code |
|-----------|-------------------|-------------|
| Xem được code của bạn | Không (bạn phải copy từng đoạn) | Có (đọc toàn bộ dự án) |
| Chỉnh sửa file | Không | Có (sửa trực tiếp trong code) |
| Chạy lệnh | Không | Có (test, cài package) |
| Hiểu ngữ cảnh | Giới hạn trong cuộc trò chuyện | Nắm bắt toàn bộ dự án |
| Kiểm tra kết quả | Bạn tự kiểm tra | Tự chạy test được |

Nói đơn giản: ChatGPT giống như nhắn tin cho một chuyên gia tư vấn. Claude Code giống như chuyên gia đó **ngồi ngay bàn làm việc** của bạn với toàn quyền truy cập vào dự án.

## CLAUDE.md chứa gì?

CLAUDE.md của dự án chúng ta bao gồm:
1. **Commands** — cách chạy dự án (`python output/chat.py`)
2. **Architecture** — cách hệ thống được tổ chức (giao diện chat, pipeline, agent)
3. **Model choices** — model nào được dùng và tại sao (Claude Sonnet, Grok-4, Opus)
4. **Database** — cách lưu trữ dữ liệu
5. **Project structure** — chức năng của từng file
6. **Framework gotchas** — các hạn chế đã biết và cách xử lý

Khi bạn yêu cầu Claude Code thực hiện thay đổi, nó đọc file này trước tiên. Nhờ đó mà nó đưa ra quyết định thông minh về việc đặt code mới Ở ĐÂU và tuân theo các pattern hiện có NHƯ THẾ NÀO.

## Quy trình phát triển với AI

Làm việc với Claude Code tuân theo 5 bước:

```
1. Tìm hiểu -> 2. Lên kế hoạch -> 3. Triển khai -> 4. Kiểm tra -> 5. Lặp lại
```

Quy trình này giống hệt agent pipeline:

| Quy trình phát triển với AI | Content Pipeline |
|----------------------------|------------------|
| Tìm hiểu vấn đề | Research Agent |
| Lên kế hoạch tiếp cận | Outline Agent |
| Triển khai giải pháp | Writer Agent |
| Kiểm tra kết quả | Human review |
| Lặp lại nếu cần | Chỉnh sửa và chạy lại |

### Các bước chi tiết:

1. **Tìm hiểu** — Claude Code dùng "plan mode" để khám phá codebase trước khi thay đổi. Giống Research Agent tìm kiếm thông tin trên web trước.
2. **Lên kế hoạch** — Claude Code thiết kế cách tiếp cận và trình bày cho bạn duyệt. Giống Outline Agent tạo cấu trúc trước khi viết.
3. **Triển khai** — Claude Code viết code, tạo file, chỉnh sửa file hiện có.
4. **Kiểm tra** — BẠN review các thay đổi dựa trên kiến thức từ khóa học này. Schema có đúng không? Model choice có phù hợp không? Có tuân theo pattern hiện có không?
5. **Lặp lại** — Nếu có gì chưa ổn, nói cho Claude Code biết cần sửa gì. Lặp lại cho đến khi đúng.

### Sub-agent

Claude Code ủy thác cho các sub-agent chuyên biệt cho từng nhiệm vụ khác nhau — giống hệt **Agno Team** trong `chat.py`! Leader (Claude Code chính) quyết định cần làm gì, sau đó ủy thác cho các chuyên gia đọc file, tìm kiếm code, chạy lệnh terminal, v.v.

## MCP — Kết nối với tài liệu

**MCP** (Model Context Protocol) server cung cấp cho Claude Code quyền truy cập vào các nguồn kiến thức bên ngoài.

Ví dụ: Dự án của chúng ta dùng **Agno framework**. Nếu không có MCP, Claude Code chỉ biết những gì có trong các file code. Khi kết nối MCP tài liệu Agno:
- Claude Code có thể tra cứu các method của Agno API
- Nó biết các tính năng và pattern mới nhất của Agno
- Nó có thể kiểm tra xem một hàm có tồn tại không trước khi sử dụng

Hiểu đơn giản, MCP là **sách tham khảo** cho Claude Code. Không có chúng, nó làm việc dựa trên trí nhớ. Có chúng, nó có thể tra cứu thông tin.

Ví dụ thực tế: "Tôi muốn thêm tool mới vào pipeline. Khi kết nối MCP tài liệu Agno, tôi chỉ cần nói 'thêm tool Google Search Console dùng pattern Agno Toolkit' và Claude Code biết chính xác API cần dùng."

## Bài tập

Điền vào chỗ trống để viết prompt hiệu quả cho Claude Code cho các nhiệm vụ sau:

1. **Dễ**: Thay đổi phong cách của writer agent từ "informative" sang "casual and conversational"
2. **Trung bình**: Thêm trường `reading_time_minutes` vào database bài viết
3. **Khó**: Thêm bước fact-checking dựa trên DuckDuckGo sau writer agent

Với mỗi prompt, hãy bao gồm: file cụ thể cần chỉnh sửa, thay đổi cần làm gì, và các ràng buộc.

````python
# Bài tập: Điền vào chỗ trống để viết prompt Claude Code hiệu quả

# 1. DỄ: Thay đổi phong cách writer agent
# Prompt tốt cần chỉ rõ: file, cần thay đổi gì, thay đổi như thế nào
easy_prompt = """
In output/agents/___, update the ___
instructions to write in a casual, conversational tone
instead of formal. Keep the existing instructions about
___ and ___ structure, just change the tone.
"""

# 2. TRUNG BÌNH: Thêm reading_time_minutes vào database
# Prompt tốt cần chỉ rõ: file, thay đổi gì, và ràng buộc
medium_prompt = """
In output/tools/___, add a new field called ___
to the Articles table. Then update the pipeline in
output/___ to calculate reading time as
word_count / 200 (average reading speed) and save it
after the ___ step completes.
"""

# 3. KHÓ: Thêm bước fact-checking sau writer
# Prompt tốt cần chỉ rõ: file, kiến trúc, pattern hiện có cần tuân theo
hard_prompt = """
Add a new ___ agent in output/agents/___
that uses ___ to verify claims in the article.
Add it as a new step in output/pipeline.py between the
___ and ___ steps. Follow the existing
error handling pattern with try/except. The agent should use
Claude ___ (same as research agent) and return a list
of claims that couldn't be verified.
"""

# In ra để review
for name, prompt in [("Dễ", easy_prompt), ("Trung bình", medium_prompt), ("Khó", hard_prompt)]:
    print(f"=== {name} ===")
    print(prompt)
    print()
````
