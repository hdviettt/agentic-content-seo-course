## Variables — Labeled Boxes

A **variable** is like a **labeled box**. You give the box a name, then put data inside.

For example:
- A box labeled `name` holds `"Viet"`
- A box labeled `age` holds `20`

The `=` sign means "assign a value" — it's not the same as "equals" in math.

````python
name = "Viet"
print(name)

age = 20
print(age)
````

## Strings and f-strings — Inserting Variables into Text

A **string** is text data, always wrapped in quotes: `"..."` or `'...'`.

An **f-string** lets you insert variable values into a string. Just add `f` before the opening quote and put variables inside `{}`.

```python
f"Hello {variable_name}"
```

This pattern is used everywhere in our project's codebase.

````python
company = "SEONGON"
role = "AI Leader"

print(f"{name} works at {company}")
print(f"Role: {role}")
print(f"{name} - {role} @ {company}")
````

## Numbers — Basic Math

Python can do math like a calculator:

| Operation      | Symbol | Example       |
|----------------|--------|---------------|
| Addition       | `+`    | `5 + 3` = 8   |
| Subtraction    | `-`    | `10 - 4` = 6  |
| Multiplication | `*`    | `3 * 7` = 21  |
| Division       | `/`    | `20 / 4` = 5.0|

You can combine numbers with f-strings for readable output.

````python
articles = 5
words_per_article = 2000
total_words = articles * words_per_article

print(f"Articles: {articles}")
print(f"Words per article: {words_per_article}")
print(f"Total words: {total_words}")
print(f"If you write {articles} articles/week, that's {articles * 4} per month!")
````
